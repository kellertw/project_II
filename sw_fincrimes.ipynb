```python
# Dependencies
import pandas as pd
```

```python
sw_casinos = pd.read_csv('data/SW_Casinos_SARStats.csv')
sw_countyloc = pd.read_csv('data/SW_countyloc.csv')
sw_casinos.head()
```

```python
sw_casinos.dtypes
```

```python
sw_casinos['Countym'] = sw_casinos['Countym'].str.split('\,+').str[0]
sw_casinos['Industry'] = sw_casinos['Industry'].str.split('\-+').str[1]

sw_casinos.head()
```

```python
sw_casinos.drop(sw_casinos.loc[sw_casinos['Suspicious Activity']=='[Total]'].index, inplace=True)
sw_casinos.drop(sw_casinos.loc[sw_casinos['Year Month']== '2020'].index, inplace=True)
del sw_casinos['Suspicious Activity']
sw_casinos.head()
```

```python
sw_casinos.count()
```

```python
sw_countyloc.head()
```

```python
sw_countyloc.dtypes
```

```python
sw_countyloc.drop(['GEOID', 'ANSICODE', 'Countym', 'ALAND', 'AWATER', 'ALAND_SQMI', 'AWATER_SQMI'], axis=1, inplace=True)

sw_countyloc.head()
```

```python
sw_countyloc.columns
```

```python
sw_county = sw_countyloc.rename(columns={"NAME":"Countym", "INTPTLAT":"Lat", "INTPTLONG                                                                                                               ":"Long" })

sw_county.head()
```

```python
sw_fincrime = pd.merge(sw_casinos, sw_county, on="Countym", how= "inner")
sw_fincrime.head()
```

```python

sw_fincrime['Count'] = sw_fincrime['Count'].str.replace(',', '').astype(float).astype(int)
```

```python
del sw_fincrime['USPS']
sw_fincrime.rename(columns={"Year Month": "Year"}, inplace=True)
sw_fincrime.head()
```

```python
licensed_df = sw_fincrime.groupby(['State', 'Year', 'Countym', 'Lat', 'Long'])['Count'].sum()
licensed_df
```

```python
casino = pd.read_csv("data/SARStats_6.csv")
casino.head()
```

```python
casino.dtypes
```

```python
casino['Count'] = casino['Count'].str.replace(',', '').astype(float).astype(int)
casino.drop(casino.loc[casino['Industry']=='[Total]'].index, inplace=True)
casino['Countym'] = casino['Countym'].str.split('\,+').str[0]
casino['Industry'] = casino['Industry'].str.split('\-+').str[1]
casino['Year Month'] = casino['Year Month'].str.split('\ +').str[0]
casino
```

```python
casino.rename(columns={"Year Month": "Year"}, inplace=True)
casino.head()
```

```python
casinos_df = pd.merge(casino, sw_county, on="Countym", how="inner")
casinos_df.head()
```

```python
del casinos_df['USPS']
casinos_df['Industry'].unique()
```

```python
tribal_df = casinos_df.loc[casinos_df["Industry"] == " Tribal Authorized Casino",:]
card_df = casinos_df.loc[casinos_df["Industry"] == " Card Club",:]
other_df = casinos_df.loc[casinos_df["Industry"] == " Other Casino/Card Club",:]
tribal_df.head()
```

```python
tribal = tribal_df.groupby(['State', 'Year', 'Countym', 'Lat', 'Long'])['Count'].sum()
tribal
```
